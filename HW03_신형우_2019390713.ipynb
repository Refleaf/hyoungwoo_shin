```python
out=[]
def iqr_outliers(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3-q1
    Lower_tail = q1 - 1.5 * iqr
    Upper_tail = q3 + 1.5 * iqr
    for i in df:
        if i > Upper_tail or i < Lower_tail:
            out.append(i)
    print("Outliers:",out)
iqr_outliers(data['MonthlyIncome'])
```

# 1. 이직 여부(Attrition) 분석

```python
# 이직 여부와 JobSatisfaction 관계
sns.countplot(x='Attrition', hue='JobSatisfaction', data = data)
plt.title('직무 만족도별 이직률')
plt.show()

# 이직 여부와 MonthlyIncome의 관계
sns.boxplot(x='Attrition', y='MonthlyIncome', data = data)
plt.title('월급에 따른 이직률')
plt.show()
```

# 2. 부서별 이직률

```python
# 부서별 이직률 계산
attrition_rate_by_department = data.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()
attrition_rate_by_department['Attrition Rate'] = attrition_rate_by_department['Yes']
print(attrition_rate_by_department)

# 부서별 이직률 시각화
attrition_rate_by_department['Attrition Rate'].plot(kind='bar', color='skyblue')
plt.title('부서별 이직률')
plt.ylabel('이직률')
plt.show()
```

# 3. 월급과 직무만족도의 관계 시각화

```python
# 월급과 직무만족도의 관계 시각화
sns.boxplot(x='JobSatisfaction', y='MonthlyIncome', data=data)
plt.title('월급 vs 직무만족도')
plt.show()
```

# 4. 근속연수와 이직 여부 분석

```python
# 근속연수와 이직 여부의 분포 - KDE 플롯
sns.kdeplot(data=data[data['Attrition'] == 'Yes']['YearsAtCompany'], label='이직자', shade=True)
sns.kdeplot(data=data[data['Attrition'] == 'No']['YearsAtCompany'], label='재직자', shade=True)
plt.title('근속연수에 따른 이직자 vs 재직자 분포')
plt.legend()
plt.show()


```
